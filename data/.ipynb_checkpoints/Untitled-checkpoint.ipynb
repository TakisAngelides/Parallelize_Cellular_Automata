```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the text file
height_profile = np.loadtxt('output.txt')
```

```python
#roughness parameter
height_mean=np.mean(height_profile)
L=len(height_profile)

w_square=(np.sum(np.square(height_profile-height_mean)))/L

print(height_profile)
print(w_square)
print("",L)
print(height_mean)
```

```python
x=np.arange(0,L)
plt.plot(x,height_profile)
```
